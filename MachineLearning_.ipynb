#1. What is a parameter?

-->A parameter is a numerical value that describes a characteristic of an entire population. Parameters are fixed values, but usually unknown, because it is difficult to collect data from the whole population.

Parameters help summarize population behavior and are often estimated using sample statistics.

Examples of parameters:

>Population mean (μ)

>Population variance (σ²)

>Population standard deviation (σ)

>Population proportion (p)

Example:

If we want to know the average height of all people in India, the true average height is a population parameter.

#2. What is correlation? What does negative correlation mean?
-->Correlation:

Correlation is a statistical measure that indicates how strongly two variables are related and in which direction they move with respect to each other.

Correlation values range from –1 to +1:

+1 → Perfect positive correlation

0 → No correlation

–1 → Perfect negative correlation

*Negative Correlation

Negative correlation means that as one variable increases, the other variable decreases.

Example:

As the price of a product increases, its demand decreases

As speed increases, travel time decreases

Negative correlation shows an inverse relationship between variables.

#3. Define Machine Learning. What are the main components of Machine Learning?
-->Definition of Machine Learning:

Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

Main Components of Machine Learning

Data

Raw information collected from various sources used to train the model.

Features

Input variables used to make predictions.

Model

A mathematical function that maps input features to output predictions.

Loss Function

Measures how incorrect the model’s predictions are.

Optimizer

Algorithm that minimizes the loss function (e.g., Gradient Descent).

Evaluation Metrics

Used to measure model performance (accuracy, precision, recall, F1-score).

#4. How does loss value help in determining whether the model is good or not?

-->The loss value quantifies the difference between actual output and predicted output.

Low loss value: Model predictions are close to actual values

High loss value: Model predictions are inaccurate

During training, the goal of a model is to minimize the loss function. A decreasing loss curve indicates that the model is learning effectively.

Example:

    In regression, Mean Squared Error (MSE) is commonly used as a loss function.

    Loss helps:

    Compare different models

    Detect underfitting or overfitting

    Improve model performance

#6. How do we handle categorical variables in Machine Learning? What are common techniques?

--->Machine learning models require numerical input, so categorical variables must be converted.

Common Techniques:

-Label Encoding

-Assigns a unique number to each category.

-One-Hot Encoding

-Creates separate binary columns for each category.

-Ordinal Encoding

-Used when categories have an order.

-Target Encoding

-Replaces categories with the mean of the target variable.

-Choosing the right encoding method depends on the model and data type.

#15. What is causation? Explain difference between correlation and causation with example.
-->Causation:

Causation occurs when one variable directly influences another.

Difference between Correlation and Causation
| Aspect      | Correlation           | Causation            |
| ----------- | --------------------- | -------------------- |
| Meaning     | Variables are related | One causes the other |
| Direction   | No cause-effect       | Direct cause-effect  |
| Reliability | Can be misleading     | Scientifically valid |

Example

    Ice cream sales and drowning deaths are correlated, but hot weather causes both, not ice cream.

Understanding this difference prevents incorrect conclusions in data analysis.
